```python
N = 2  # number of qubits
M = 23 # number of columns
matrix = [['   ' for _ in range(M)] for _ in range(N)]

for i in range(N):
    matrix[i][0] = f'\\lstick{'{$'}q_{i}${'}'}'


matrix[0][1] = f'\\gate{'{\\gateg{Ry}[x_'}{0}{']}'}'

for i in range(3,10,2):
    j = 1
    matrix[1][i] = f'\\gate{'{\\gateg{Ry}[x_'}{j}{']}'}'
    j += 1
    if i == 5:
        matrix[0][i] = '\\gate{\\gateg{X}}'
    elif i == 9:
        matrix[0][i] = '\\gate{\\gateg{X}}\\slice{}'

for i in range(2,9,2):
    matrix[0][i] = '\\ctrl{1}'
    matrix[1][i] = '\\targ{}'

# END encoding
# ANsatz

w = 0
for i in range(10,21,2):
    angles1 = f"w_{'{'}{w}{'}'},w_{'{'}{w+1}{'}'},w_{'{'}{w+2}{'}'}"
    w += 3
    angles2 = f"w_{'{'}{w}{'}'},w_{'{'}{w+1}{'}'},w_{'{'}{w+2}{'}'}"
    w += 3
    matrix[0][i] = f'\\gate{'{\\gateg{R3}_{'}{angles1}{'}}'}'
    matrix[1][i] = f'\\gate{'{\\gateg{R3}_{'}{angles2}{'}'}'
    matrix[0][i+1] = '\\ctrl{1}'
    matrix[1][i+1] = '\\targ{}'
    

matrix[0][22] = '\\meter{}'
matrix[1][22] = '\\ground{}'

for row in matrix:
    print(' & '.join(row) + r' \\')
```

```python
N = 4 # number of qubits
M = 17 # number of columns
matrix = [['' for _ in range(M)] for _ in range(N)]



counter = 0
for i in range(N):
    matrix[i][counter] = f'\\lstick{'{$'}q_{i}${'}'}'
counter += 1

for i in range(N):
    matrix[i][counter] = '\\gate{\\gateg{H}}'
counter += 1
c = 0
for i in range(4):
        # For odd-indexed features, use Z-rotation in the angle embedding
    r = 'z' if i % 2 else 'x'
    for j in range(N):
        matrix[j][counter] = f'\\gate{'{\\gateg{R'}{r}{'}[x_{'}{c}{'}]}'}'
        c += 1
    counter += 1
        
w = 0
for i in range(N):
    matrix[i][counter] = f'\\gate{'{\\gateg{Ry}_{w_{'}{w}{'}}}'}'
    w += 1
counter += 1

matrix[N-1][counter] = f'\\ctrl{'{-3}'}'
matrix[0][counter] = f'\\targ{'{}'}'
counter += 1
for i in range(N-1):
    matrix[i][counter] = f'\\ctrl{'{1}'}'
    matrix[i+1][counter] = f'\\targ{'{}'}'
    counter += 1

for i in range(N):
    matrix[i][counter] = f'\\gate{'{\\gateg{Ry}_{w_{'}{w}{'}}}'}'
    w += 1
counter += 1

for i in range(N-1):
    matrix[i][counter] = f'\\ctrl{'{1}'}'
    matrix[i+1][counter] = f'\\targ{'{}'}'
    counter += 1
matrix[N-1][counter] = f'\\ctrl{'{-3}'}'
matrix[0][counter] = f'\\targ{'{}'}'
counter += 1

matrix[0][counter] = '\\meter{}'
for i in range(1, N):
    matrix[i][counter] = '\\ground{}'


for row in matrix:
    print(' & '.join(row) + r' \\')
```
